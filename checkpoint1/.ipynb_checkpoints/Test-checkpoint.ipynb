```python
def fix_strings(df,col):
    
    for i in range(0,len(df)):
        for j,x in enumerate(df[col][i]):
            if x.isalpha() == False and x != ' ':
                df[col][i] = df[col][i].replace(x, '')
    
    return df
```

```python
def fill_iron(df):

    for i in range(0,len(df)):
        
        if pd.isna(df['Iron (% DV)'][i]) == True:
            df['Iron (% DV)'][i] = 0
            df['Iron (% DV)'][i] = int(df['Iron (% DV)'][i])
    
    return df
```

```python
def standardize_names(df):
    col_name = {}
   
    for col in df.columns:
        string = col.lower()
        for j,x in enumerate(string):
            if x == '(':
                string = string[0:j - 1]
        col_name[col] = string
    
    df = df.rename(columns = col_name)
            
    return df
```

```python
def fix_caffeine(df):
    k = df['Caffeine (mg)']
    for i,x in enumerate(k):
        
        if x == 'varies' or x == 'Varies':
            k = k.drop(index = i)
        
        else:
            k[i] = float(k[i])
            
    k = k.fillna(round(k.mean(),0))
    df['Caffeine (mg)'] = k.astype('int64')
    
    return df
```

```python
fix_caffeine(df)
```
